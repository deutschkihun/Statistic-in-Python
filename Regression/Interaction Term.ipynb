## Interaction term

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

df = pd.read_csv('house_prices.csv')
df.head()
```

```python
df['area_bed'] = df['area']*df['bedrooms']
```

```python
df['intercept']=1
lm =sm.OLS(df['price'],df[['intercept', 'bedrooms','area_bed']])
result = lm.fit()
result.summary()
```
